```python
import os
import tifffile as tff
import numpy as np
from skimage.io import imsave
```

```python
imgPath = '/Users/zhengj10/Desktop/Positron-ST&excitability/raw_data'
```

```python
def getPathInfo(path):
    filePath = [d for d in os.listdir(path) if os.path.isdir(os.path.join(path, d))]
    filePath = sorted(filePath)
    return filePath
```

```python
datePath = getPathInfo(imgPath)
datePath
```

```python
for date in datePath:
    platePath = getPathInfo(os.path.join(imgPath, date))
    for plate in(platePath):
        cellPath = getPathInfo(os.path.join(imgPath, date, plate))
        for cell in cellPath:
            img = os.path.join(imgPath, date, plate, cell, '128x128.tif')
            os.path.join(imgPath, date, plate, cell)
            if not os.path.exists(img):
                continue
            img = tff.TiffFile(img)
            img = img.asarray()
            img = np.mean(img, axis=0)
            imsave(os.path.join(imgPath, date, plate, cell)+'/128x128_mean.jpg',img)
```
