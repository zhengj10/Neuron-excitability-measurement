### Swarmplot + pointplot

```python
params = ['Rm', 'Cm', 'RMP', 'uIth', 'Vth', 'AP_amplitude', 'AP_width', 'AHP_amplitude', 'AHP_width', 'No_APs_1.5x']
labels = ['Rm (Mohm)', 'Cm (pF)', 'RMP (mV)', 'Ith (pA/pF)', 'Vth (mV)', 'AP_amplitude (mV)', 'AP_width (ms)', 
          'AHP_amplitude (mV)', 'AHP_width (ms)', 'No_APs at 1.5x threshold']
```

```python
fig = plt.figure(figsize = (16,7))

for counter, value in enumerate(params):
    if counter < 4:
        ax  = fig.add_subplot(2,5,counter+1)
        s_bar = sns.swarmplot(x="Plasmid", y=value, hue="Transfected", hue_order=[True,False],
            data=df, ax = ax, size=8, dodge=True, alpha=0.3)
        s_point = sns.pointplot(x="Plasmid", y=value, hue="Transfected", hue_order=[True,False],
            data=df, ax = ax, size=8, linestyles=["",""], dodge =0.4, capsize=0.2)
        s_bar.legend_.remove()
        s_bar.set_xlabel('')
        s_bar.set_xticklabels('')
        s_bar.set_ylabel(labels[counter], size=13)
        
    elif counter == 4:
        ax  = fig.add_subplot(2,5,counter+1)
        s_bar = sns.swarmplot(x="Plasmid", y=value, hue="Transfected", hue_order=[True,False],
            data=df, ax = ax, size=8, dodge=True, alpha=0.3)
        s_point = sns.pointplot(x="Plasmid", y=value, hue="Transfected", hue_order=[True,False],
            data=df, ax = ax, size=8, linestyles=["",""], dodge =0.4, capsize=0.2)
        s_bar.legend_.remove()
        s_bar.set_xticklabels('')
        s_point.set_xlabel('')
        s_bar.set_ylabel(labels[counter], size=13)
        hand, labl = ax.get_legend_handles_labels()
        hd = [h for h in hand] 
        L = ax.legend(hd[0:2], labl[0:2], loc='center left', bbox_to_anchor=(1, 0.75))
        L.get_texts()[0].set_text('+')
        L.get_texts()[1].set_text('-')
        
    else:
        ax  = fig.add_subplot(2,5,counter+1)
        s_bar = sns.swarmplot(x="Plasmid", y=value, hue="Transfected",hue_order=[True,False],
            data=df, ax = ax, size=8, dodge=True, alpha=0.3)
        s_point = sns.pointplot(x="Plasmid", y=value, hue="Transfected",hue_order=[True,False],
            data=df, ax = ax, size=8, linestyles=["",""], dodge =0.4, capsize=0.2)     
        s_bar.legend_.remove()
        s_bar.set_xlabel('')
        s_bar.set_ylabel(labels[counter], size=13)
        s_bar.set_xticklabels(s_bar.get_xticklabels(), rotation=45, ha = 'right', size=13)

plt.tight_layout()
# plt.savefig('/Users/zhengj10/Desktop/swarm+point_32x14.png', dpi=600)
plt.show()
```

### Statistics

```python
from scipy.stats import mannwhitneyu
```

```python
p_val = {}
plasmids = [65.1203,65.1204]
des = {}
for plasmid in(plasmids):
    col_1 = str(plasmid) + '+'
    col_2 = str(plasmid) + '-'
    col_3 = str(plasmid) + '_pval'
    p_val[col_1] = {}
    p_val[col_2] = {}
    p_val[col_3] = {}
    des[plasmid] = {}
    for param in(params):

        group_1 = df[(df.Plasmid == plasmid)&(df.Transfected==1)]
        group_2 = df[(df.Plasmid == plasmid)&(df.Transfected==0)]

        des[plasmid]['+'] = len(group_1)
        des[plasmid]['-'] = len(group_2)
        if param == 'Rm':
            print(str(plasmid) + '+ = ' + str(des[plasmid]['+']))
            print(str(plasmid) + '- = ' + str(des[plasmid]['-']))
        a = group_1[param]
        b = group_2[param]
        a_mean = np.mean(a)
        a_std = np.std(a)
        a_mean = round(a_mean,2)
        a_std = round(a_std,2)
        b_mean = np.mean(b)
        b_std = np.std(b)
        b_mean = round(b_mean,2)
        b_std = round(b_std,2)
        stat, p = mannwhitneyu(a, b, alternative='two-sided')
        p = round(p,2)
        p_val[col_1][param] = str(a_mean) + '+-' + str(a_std)
        p_val[col_2][param] = str(b_mean) + '+-' + str(b_std)
        p_val[col_3][param] = p
```

```python
df_pval = pd.DataFrame(p_val)
df_pval = df_pval.transpose()
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_colwidth', 20)
col_sel = ['Rm','Cm','RMP','uIth','AP_amplitude','AP_width']
df_pval = df_pval[col_sel]
df_pval.index = [['Positron','Positron','Positron','Positron-ST','Positron-ST','Positron-ST'],
                ['Positron+ (n = %d)' %des[65.1203]['+'],'Positron- (n = %d)' %des[65.1203]['-'],'P',
                 'Positron-ST+ (n = %d)' %des[65.1204]['+'],'Positron-ST- (n = %d)' %des[65.1204]['-'],'P']]
df_pval
# df_pval.to_excel(dataPath + '/stat_table.xlsx')
```
