```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('/Users/zhengj10/Desktop/GENIE/Data/Voltron-ST&Kv2.1/summary_VI.xlsx', sheet_name='Summary')
df
```

## 476

```python
mean_col = ['476_control_non_transfected', '476_control_transfected', '476_GxTx_non_transfected', '476_GxTx_transfected']
err_col = ['sem_1', 'sem_2', 'sem_3', 'sem_4']
fmt = ['s','s','o','o']
fillstyle = ['none', 'full', 'none', 'full']
legend = ['Control_non_transfected (6)', 'Control_transfected (7)', 'GxTx_non_transfected (5)', 'GxTx_transfected (5)']
```

```python
fig = plt.figure(figsize = (10,6))
x = df['Voltage'].values
for i in range(4):
    y = df[mean_col[i]].values
    e = df[err_col[i]].values
    plt.errorbar(x, y, yerr=e, fmt = fmt[i], fillstyle = fillstyle[i]  , color = 'k', label = legend[i], markersize=10)
plt.legend(loc='upper left')
plt.xlabel('mV', size=12)
plt.ylabel('pA/pF', size=12)
plt.show()
fig.savefig('/Users/zhengj10/Desktop/476.png')
```
